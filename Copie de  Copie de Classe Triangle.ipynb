### A - Coder une classe **Triangle** contenant 3 points **A**, **B** et **C**, un **constructeur** qui initialise ce triangle avec 3 points de coordonnées (0, 0), une méthode **perimetre()** qui renvoie la somme des longueurs des segments AB, BC et CA, et une méthode **barycentre()** qui renvoie un Point **G** avec les coordonnées du centre de gravité du triangle ABC.

```python
import math
#pour   afficher les different point du triangle
class Point:
   def __init__(self):
      self.x = 0
      self.y = 0
   def deplacer(self, deltaX, deltaY):
      self.x += deltaX
      self.y = self.y + deltaY
   def afficher(self):
    print("x = ", self.x)
    print(f"y = {self.y}")
```

```python
import math
#calculer les longueur des segment
class Segment:
  def __init__(self, A = Point(), B = Point()):
    self.A = A
    self.B = B
  def longueur(self):
    res = (self.A.x - self.B.x)**2 + (self.A.y - self.B.y)**2
    return math.sqrt(res)
  def milieu(self):
    M = Point()
    M.x = (self.B.x + self.A.x) / 2
    M.y = (self.B.y + self.A.y) / 2
    return M


```

```python
import math
#cree le triangle
class Triangle:
  def __init__(self,  A = Point(), B = Point(), C = Point()):
      self.A = A
      self.B = B
      self.C = C
      self.AB = Segment(self.A,self.B)
      self.BC = Segment(self.B,self.C)
      self.CA = Segment(self.C,self.A)
#calcul du perimetre du triangle
  def perimetre(self):
        return self.AB.longueur() + self.BC.longueur() + self.CA.longueur()

  def barycentre(self):
      xG = (self.A.x + self.B.x + self.C.x) / 3
      yG = (self.A.y + self.B.y + self.C.y) / 3
      #creation du point G
      G = Point()
      #deplacement du point G pour le mettre au barycentre
      G.deplacer(xG,yG)

      return G


  #def __init__(self, x=0, y=0):
      #self.x = x
      #self.y = y

      #print("Barycentre G :", G_x, G_y)









```

```python
import math
#creation des different point present sur la figure
A = Point()
B = Point()
C = Point()
A.deplacer(3, 2)
B.deplacer(7, 5)
C.deplacer(5, 8)

ABC = Triangle(A,B,C)

print(ABC.perimetre())

Ap = ABC.BC.milieu()
Bp = ABC.CA.milieu()
Cp = ABC.AB.milieu()

G = ABC.barycentre()

print("Coordonnées de G :", G.x, G.y)
```
